```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Step 2: Load Dataset
df = pd.read_csv('../data/train.csv')

print("✅ Dataset loaded successfully!")
print("Shape:", df.shape)
df.head()
```

```python
# Step 3: Basic Overview
print("----- Dataset Info -----")
df.info()

print("\n----- Summary Statistics -----")
df.describe().T
```

```python
# Step 4: Target Variable Distribution
sns.countplot(data=df, x='Credit_Score', palette='Set2')
plt.title("Distribution of Credit Score")
plt.show()

df['Credit_Score'].value_counts()
```

```python
# Step 5: Check Missing Values
print("----- Missing Values -----")
df.isnull().sum().sort_values(ascending=False).head(15)
```

```python
# Step 6: Handle Missing Values
# Fill numeric NaN values with mean
num_cols = df.select_dtypes(include=['float64','int64']).columns
for col in num_cols:
    df[col] = df[col].fillna(df[col].mean())

# Fill categorical NaN values with mode
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print(" Missing values handled successfully!")
df.isnull().sum().sum()
```

```python
# Step 7: Fix Mixed Data Types (Monthly_Balance)
df['Monthly_Balance'] = pd.to_numeric(df['Monthly_Balance'], errors='coerce')
df['Monthly_Balance'].fillna(df['Monthly_Balance'].mean(), inplace=True)
print("Fixed mixed-type column: Monthly_Balance")
```

```python
# Step 8: Encode Categorical Columns
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].astype(str)   # ensure uniform type
    df[col] = le.fit_transform(df[col])

print(" Encoded all categorical columns!")
df.info()
```

```python
# Step 9: Save Cleaned Dataset
df.to_csv('../data/cleaned_credit_score.csv', index=False)
print(" Cleaned dataset saved successfully at ../data/cleaned_credit_score.csv")
```
